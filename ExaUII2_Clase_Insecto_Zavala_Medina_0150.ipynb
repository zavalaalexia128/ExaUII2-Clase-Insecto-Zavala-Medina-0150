#### Crear la clase "Insecto", con 3 atributos (nombre, habitat, tieneAlas)
#### 2 funciones leerDatos(), mostrarDatos() y 2 objetos (insecto1, insecto2).
#### que sea interctivo --> input
#### no usar while true, if, for, lista y diccionarios

### Insectos
![](https://raw.githubusercontent.com/zavalaalexia128/ExaUII2-Clase-Insecto-Zavala-Medina-0150/refs/heads/main/Claseinsctos.png)

```python

class Insecto:
    def __init__(self, nombre=None, habitat=None, tieneAlas=None):
        self.nombre = nombre
        self.habitat = habitat
        self.tieneAlas = tieneAlas

    def leerDatos(self):
        self.nombre = input("Ingrese el nombre del insecto: ")
        self.habitat = input("Ingrese el hábitat del insecto: ")
        self.tieneAlas = input("¿Tiene alas? (Si/No): ")
        return self.nombre, self.habitat, self.tieneAlas

    def mostrarDatos(self):
        print(f"Nombre: {self.nombre}")
        print(f"Hábitat: {self.habitat}")
        print(f"Tiene alas: {self.tieneAlas}")
        return self.nombre, self.habitat, self.tieneAlas

# --- Registro de Objetos ---

print("--- Registro de Insecto Numero 1 ---")
insecto1 = Insecto()
insecto1.leerDatos()
insecto1.mostrarDatos()
print(". ݁₊ ⊹ . ݁˖ . ݁")

print("--- Registro de Insecto Numero 2 ---")
insecto2 = Insecto()
insecto2.leerDatos()
insecto2.mostrarDatos()
print(". ݁₊ ⊹ . ݁˖ . ݁")

print("Software realizado por Zavala Medina Alexia Paola NL = 54")
```
